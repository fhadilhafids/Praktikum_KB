Import Library / Modul yang diperlukan

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Membaca Dataset

```python
data=pd.read_csv("Dataset-SA.csv")
```

MAAF BANG, KARENA TIPE DATANYA BERBEDA DENGAN YANG DI KAGGLE JADI SAYA UBAH TIPE DATANYA BIAR CODENYA BERHASIL

```python
data = data[data["product_price"].str.isnumeric()]
data["product_price"] = data["product_price"].astype("int64")
```

```python
data = data[data["Rate"].str.isnumeric()]
data["Rate"] = data["Rate"].astype("int64")
```

Melihat informasi dari dataset

```python
data.info()
```

1. Grafik Batang

Dari grafik dibawah dapat disimpulkan konsumen dengan feedback positif memberi rating yang tinggi. Sedangkan, konsumen dengan feedback negative cenderung memberi rating dibawah 2. Untuk konsumen dengan feedback netral memberi rating diatas 3

```python
x = data['Sentiment']
y = data['Rate']
colors = ['green', 'red', 'yellow']
sns.barplot(
    x=x,
    y=y,
    palette=colors
)
plt.xlabel('Sentiment')
plt.ylabel('Rate')
plt.title('Grafik Tingkat Kepuasan Konsumen Berdasarkan Rating')
plt.show()
```

2. GRAFIK PIECHART

Dari grafik dibawah dapat disimpulkan banyak konsumen memberi rating 5 pada product yaitu sebanyak 57,9% . Sedangkan, konsumen cenderung sedikit memberi rating 2 pada product yaitu sebanyak 3.2%.

```python
data = data.dropna(subset=['Rate'])
data_pie = data['Rate'].value_counts().rename_axis('Rate').reset_index(name='Sentiment')

plt.figure(figsize=(10, 10))
plt.pie(
    data_pie['Sentiment'], 
    labels=data_pie['Rate'],
    startangle=90,
    autopct='%.1f%%'
)
plt.xlabel('Sentiment')
plt.ylabel('Rate')
plt.title('Data Rating Konsumen')
plt.legend(data_pie['Rate'], title='Rate', loc='upper left')
plt.show()
```

3. GRAFIK HEATMAP

Pada grafik dibawah ini korelasi antara atribut numerik 'product_price' dan 'Rate' dalam data. Heatmap menunjukkan seberapa kuat hubungan korelasi antara 'product_price' dan 'Rate'. Nilai korelasi berkisar antara -1 hingga 1, di mana -1 menunjukkan korelasi negatif sempurna, 0 menunjukkan tidak ada korelasi, dan 1 menunjukkan korelasi positif sempurna.

```python
atribut_numerik = ['product_price','Rate']

correlation_matrix = data[atribut_numerik].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.xlabel('Product Price')
plt.ylabel('Rate')
plt.title('Heatmap Korelasi Product Price dan Rate')
plt.show()
```

4. GRAFIK BOXPLOT

Pada Grafik Box Plot dibawah konsumen yang Feedbacknya Positive memberi rating di rentang 3 - 5 dan yang Negative memberi rating di rentang 1 - 3. Sedangkan yang Netral memberi rating di rentang 2 - 5.

```python
palette = "Set2"
plt.figure(figsize=(10,6))
sns.boxplot(
    data   = data,
    x      = 'Sentiment',
    y      = 'Rate',
    palette=palette
)

plt.xlabel('Sentiment')
plt.ylabel('Rate')
plt.title('Grafik Tingkat Kepuasan Konsumen Berdasarkan Rating')
plt.show()
```

5. Grafik Histogram

Dari grafik dibawah dapat disimpulkan banyak konsumen memberi rating 5 pada product. Sedangkan, konsumen cenderung sedikit memberi rating 2 pada product.

```python
plt.hist(data['Rate'], bins=20, color='yellow')
plt.title('Histogram Rating')
plt.ylabel('Frekuensi kemunculan')
plt.xlabel('Rating')
plt.show()
```
